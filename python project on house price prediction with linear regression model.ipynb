```python
#Encoding technique
#status 
from sklearn import preprocessing
my_status=preprocessing.LabelEncoder()
df['Status']=my_status.fit_transform(df['Status'])  #1:"Ready_to_move"   0:'Almost_ready'

#Furnishing
from sklearn import preprocessing
my_furnishing=preprocessing.LabelEncoder()
df['Furnishing']=my_furnishing.fit_transform(df['Furnishing']) #0:Furnished,1:semi-furnished,2:unfurnished


#transaction
from sklearn import preprocessing
my_transaction=preprocessing.LabelEncoder()
df['Transaction']=my_transaction.fit_transform(df['Transaction'])  #1:'Resale'  0:'New_Property'




```

```python
#type
df["Type"] = df["Type"].astype("category")
df["Type"] = df["Type"].cat.codes  #2:'Builder-floor'  1:'Apartment'
```

```python
# dividing in independent and dependent variables
X=df.iloc[:,[0,1,2,11]]
X
```

```python
Y=df.iloc[:,[12]]
Y
```

```python
#splitting in testing and training set
from sklearn.model_selection import train_test_split
x_train , x_test , y_tain, y_test = train_test_split(X,Y,test_size = 0.2,random_state = 0)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler    
sc= StandardScaler() 

x_train.iloc[:,:] = sc.fit_transform(x_train.iloc[:,:])
print(x_train)

```

```python
# standardisation for testing data

x_test.iloc[:,:] = sc.transform(x_test.iloc[:, :])
print(x_test)
```

```python
# Now apply algorithm-
#Linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Prediction of Test and Training set result  
y_pred= lr.predict(x_test)  
x_pred= lr.predict(x_train)  
```

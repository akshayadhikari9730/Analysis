```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df.head()
```

```python
sns.lineplot(x="Continent",y="Rank",data=df,color="g")
plt.show()
```

```python
sns.set (style="whitegrid")
sns.barplot(x=df["Country"].head(),y=df["2022 Population"].head(),color="r",palette="rocket")
plt.show()
```

```python
sns.set (style="whitegrid")
sns.scatterplot(x=df["Country"].head(3),y=df["2022 Population"].head(3),color="r")
plt.title("scatter plot showing 2022 population of first 3 countries")
plt.show()
```

```python
sns.histplot(df["Growth Rate"],kde=True,color="g")
plt.show()
```

```python
sns.displot(df, x="Growth Rate", alpha=.4, rug=True)
```

```python
sns.jointplot(x="World Population Percentage",y="Growth Rate",color="olive",kind="reg",data=df)
plt.show()
```

```python
sns.boxplot(x="Continent",y="Growth Rate",data=df,palette="Set1",linewidth=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.scatter(x="Country",y="Area (km²)")
plt.show()

plt.subplot(2,1,2)
plt.scatter(x="Country",y="Density (per km²)")
plt.show()
```

```python
df["Growth Rate"].head(3)
plt.violinplot(df["Growth Rate"].head(3),showmedians=True)
```

```python
x=df["Growth Rate"].head()
y=df["Continent"].head()
plt.pie(x,labels= y,autopct='%0.1f%%',colors=["red","blue","grey","yellow","black"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
cor=df.corr(method="pearson")
fig,ax=plt.subplots(figsize=(10,5))
sns.heatmap(cor,annot=True,cmap='YlGnBu',ax=ax)
```

```python
c=df.groupby("Continent")[['Growth Rate']].mean()
plt.rcParams['figure.figsize']=(12,7)
plt.bar(c.index,c['Growth Rate'],color='red',width=0.6)
plt.xticks(rotation=35,ha='right')
```

```python
sns.set_style('darkgrid')
plt.rcParams['figure.figsize']=(15,7)
plt.title("Continent vs Growth rate")
sns.scatterplot(x=df['Continent'],y=df['Growth Rate'],s=200)
plt.legend(loc='upper left',fontsize=11)
plt.xlabel("Continent")
plt.ylabel("Growth Rate")
```
